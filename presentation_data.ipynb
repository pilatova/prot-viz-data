```python
#code that was taken from lowest_highest_frequency (4).ipynb
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/pilatova/prot-viz-data/test/dist2.tsv"

dist = pd.read_csv(
    url,
    sep='\t',
    header=0,
    index_col=0
)

melted = pd.DataFrame()
proteins = []
pairs = []
vals = []
for prot in dist.index:
    for pair in dist.columns:
        proteins.append(prot)
        pairs.append(pair)
        vals.append(dist.loc[prot, pair])
melted['protein_id'] = proteins
melted['value'] = vals
melted['pair'] = pairs

#code made just for better visualization, for this tables are used only proteins that we have in our presentation
def get_min_with_counts(group, top_n=100):
    group_df = melted[melted['protein_id'].isin(group)]
    sorted_df = group_df.sort_values(['protein_id', 'value'], ascending=[True, True])
    min_values = sorted_df.groupby('protein_id').head(top_n)
    counts = min_values['protein_id'].value_counts().reset_index()
    counts.columns = ['protein_id', 'count']
    min_extreme = min_values.groupby('protein_id')['value'].min().reset_index()
    result = pd.merge(min_extreme, counts, on='protein_id')
    return result.sort_values('count', ascending=False).reset_index(drop=True)

def get_max_with_pairs(group):
    return (melted[melted['protein_id'].isin(group)]
            .sort_values(['protein_id', 'value'], ascending=[True, False])
            .groupby('protein_id').head(1)
            .reset_index(drop=True))

#  a table for P60002, P60059, P62917 (the smallest values)
group1 = ['P60002', 'P60059', 'P62917']
min_group1 = get_min_with_counts(group1)
display(min_group1)

# a table for Q6UXN9, P59998, Q7RTV0 (the smallest values)
group2 = ['Q6UXN9', 'P59998', 'Q7RTV0']
min_group2 = get_min_with_counts(group2)
display(min_group2)

# a table for P26951, Q8N5Q1, Q6ZQX7 (the biggest values)
group3 = ['P26951', 'Q8N5Q1', 'Q6ZQX7']
max_group3 = get_max_with_pairs(group3)
display(max_group3)
```
